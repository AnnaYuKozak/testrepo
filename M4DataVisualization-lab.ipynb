# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# Step 1: Import the necessary modules
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Connect to the database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 3: Run an SQL query to extract the "ConvertedComp" column
QUERY = """
SELECT ConvertedComp 
FROM master
WHERE ConvertedComp IS NOT NULL
"""

# Step 4: Load the data into a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Step 5: Plot the histogram
plt.figure(figsize=(10, 6))
df['ConvertedComp'].plot(kind='hist', bins=30, edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
# Step 1: Import the necessary modules
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Connect to the database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 3: Run an SQL query to extract the "Age" column
QUERY = """
SELECT Age
FROM master
WHERE Age IS NOT NULL
"""

# Step 4: Load the data into a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Step 5: Plot the box plot
plt.figure(figsize=(8, 6))
df.boxplot(column=['Age'])
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.grid(True)
plt.show()

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
# Step 1: Import the necessary modules
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Connect to the database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 3: Run an SQL query to extract "Age" and "WorkWeekHrs" columns
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
WHERE Age IS NOT NULL AND WorkWeekHrs IS NOT NULL
"""

# Step 4: Load the data into a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Step 5: Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.grid(True)
plt.show()

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
# Step 1: Import the necessary modules
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Connect to the database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 3: Run an SQL query to extract "WorkWeekHrs", "CodeRevHrs", and "Age" columns
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE WorkWeekHrs IS NOT NULL AND CodeRevHrs IS NOT NULL AND Age IS NOT NULL
"""

# Step 4: Load the data into a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Step 5: Plot the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*10, alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs (Bubble size = Age)')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.grid(True)
plt.show()

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# Step 1: Import the necessary modules
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Connect to the database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 3: Run an SQL query to extract the data from the DatabaseDesireNextYear table
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""

# Step 4: Load the data into a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Step 5: Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['Count'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
# Step 1: Import the necessary modules
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Connect to the database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 3: Run an SQL query to extract WorkWeekHrs and CodeRevHrs for the age group 30 to 35
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35 AND WorkWeekHrs IS NOT NULL AND CodeRevHrs IS NOT NULL
"""

# Step 4: Load the data into a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Step 5: Calculate the median values
median_work_week_hrs = df['WorkWeekHrs'].median()
median_code_rev_hrs = df['CodeRevHrs'].median()

# Step 6: Prepare the data for the stacked bar chart
data = {
    'MedianWorkWeekHrs': [median_work_week_hrs],
    'MedianCodeRevHrs': [median_code_rev_hrs]
}

df_median = pd.DataFrame(data, index=['Age 30-35'])

# Step 7: Plot the stacked bar chart
df_median.plot(kind='bar', stacked=True, figsize=(8, 6), color=['#1f77b4', '#ff7f0e'])
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs (Age 30-35)')
plt.ylabel('Hours')
plt.xticks(rotation=0)  # Keep the x-axis labels horizontal
plt.legend(['Median Work Week Hours', 'Median Code Review Hours'])
plt.grid(True)
plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
# Step 1: Import the necessary modules
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Connect to the database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 3: Run an SQL query to extract ConvertedComp and Age for ages 45 to 60
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 25 AND 30 AND ConvertedComp IS NOT NULL
"""

# Step 4: Load the data into a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Step 5: Calculate the median ConvertedComp for each age
median_comp_by_age = df.groupby('Age')['ConvertedComp'].median()

# Step 6: Plot the line chart
plt.figure(figsize=(10, 6))
median_comp_by_age.plot(kind='line', marker='o')
plt.title('Median ConvertedComp for Ages 25 to 30')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.grid(True)
plt.show()

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
# Step 1: Import the necessary modules
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Connect to the database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 3: Run an SQL query to extract the MainBranch data
QUERY = """
SELECT MainBranch, COUNT(*) as Count
FROM master
GROUP BY MainBranch
"""

# Step 4: Load the data into a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Step 5: Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
df.plot(kind='barh', x='MainBranch', y='Count', color='skyblue', legend=False)
plt.title('Distribution of MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.grid(True)
plt.show()

```
